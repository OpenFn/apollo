````python
evaluate_answer_template = """
For the given Question and Answer pair, evaluate whether the answer at least briefly answers the question (True), or if it entirely fails to answer the question (False).

Additionally, check the content of the `answer` variable for the presence of the pattern `)(state)` using regex. If this pattern is found in the `answer`, consider the response invalid and mark it as `Success: False`.

<example>
Q: "How can I map CommCare case data to create multiple FHIR patients in a single job?"
A: "get('Patient)"
Regex Match: False
Success: True

Q: "How can I map CommCare case data to create multiple FHIR patients in a single job?"
A: "fn(state => {{ return {{ ...state, people: [] }}; }})(state);"
Regex Match: True
Success: False

Now, following Q/A pair, by ONLY outputting "True" or "False":
Q: {question}
A: {answer}
Regex Match: 
Success:
"""
````

````python
# Evaluate QA pairs

results = []

for q, a, version in qa_pairs:
    evaluate_answer_template_filled = evaluate_answer_template.format(question=q, answer=a)

    message = client.messages.create(
        model="claude-3-5-sonnet-20241022",
        max_tokens=1000, #TODO limit to 1
        temperature=0,
        system="You are an AI programming assistant. Follow the user's requirements carefully and to the letter.",
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": evaluate_answer_template_filled
                    }
                ]
            }
        ]
    )
    result = message.content[0].text
    print(result)
    results.append([q, a, version, result])
````

````python
import pandas as pd
import os
import re

def extract_success(result_text):
    # Use regex to extract "Success: True" or "Success: False"
    import re
    match = re.search(r'Success:\s*(True|False)', result_text)
    return match.group(1) if match else None

def extract_code(answer_text):
    # Regex to match code inside ```javascript ... ```
    match = re.search(r"```javascript\s*([\s\S]*?)\s*```", answer_text)
    return match.group(1).strip() if match else None


results_df = pd.DataFrame(results, columns=["question", "answer", "system_prompt_version", "result"])
results_df['code'] = results_df['answer'].apply(extract_code)
results_df['result'] = results_df['result'].apply(extract_success).map({'True': True, 'False': False})
v1_mean = results_df[results_df['system_prompt_version'] == 'v1']['result'].mean()
v2_mean = results_df[results_df['system_prompt_version'] == 'v2']['result'].mean()

print(f"V1 success rate: {v1_mean}, V2 success rate: {v2_mean}")
downloads_path = os.path.join(os.path.expanduser("~"), "Downloads", "202411_claude_system_prompt_issue_97.csv")
results_df.to_csv(downloads_path, index=False)

results_df
````
